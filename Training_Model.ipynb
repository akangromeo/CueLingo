**Prepare The dataset**

```python
train_dataset_path = "/content/drive/MyDrive/DatasetBangkit/DatasetV2/Train/images"
validation_dataset_path = "/content/drive/MyDrive/DatasetBangkit/DatasetV2/Val/images"
```

**Load The Dataset**

> Indented block



```python
train_data = object_detector.Dataset.from_pascal_voc_folder(
    '/content/drive/MyDrive/DatasetBangkit/DatasetV2/Train',
    cache_dir="/content/drive/MyDrive/DatasetBangkit/DatasetV2/Train/cache",
)

val_data = object_detector.Dataset.from_pascal_voc_folder(
    '/content/drive/MyDrive/DatasetBangkit/DatasetV2/Val',
    cache_dir="/content/drive/MyDrive/DatasetBangkit/DatasetV2/Val/cache",
)
```

**Training The model**

Set Batch_size = 24
set Learning_rate = 0.1
Set Epochs = 55



```python
hparams = object_detector.HParams(batch_size=24, learning_rate=0.1, epochs=55, export_dir='exported_model')
options = object_detector.ObjectDetectorOptions(
    supported_model=object_detector.SupportedModels.MOBILENET_V2,
    hparams=hparams
)
model = object_detector.ObjectDetector.create(
    train_data=train_data,
    validation_data=val_data,
    options=options)
```

**Model Performance Evaluation**

```python
loss, coco_metrics = model.evaluate(val_data, batch_size=4)
print(f"Validation loss: {loss}")
print(f"Validation coco metrics: {coco_metrics}")
```

**Export Model To Tflite**

```python
model.export_model('cuelingo.tflite')
!ls exported_model
files.download('exported_model/cuelingo.tflite')
```
